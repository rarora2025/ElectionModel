```python
import operator as op

c1list = list(newdf.candidate1)
c2list = list(newdf.candidate2)
c3list = list(newdf.candidate3)

def findCount(cand):
    
    return (int(op.countOf(c1list, cand)) + int(op.countOf(c2list, cand)) + int(op.countOf(c3list, cand)))-1
    
c1_incumbency = []
c2_incumbency = []
c3_incumbency = []
for x in range(len(newdf)):
    c1_incumbency.append(findCount(str(newdf.iloc[[x]].candidate1[x])))
    c2_incumbency.append(findCount(str(newdf.iloc[[x]].candidate2[x])))
    c3_incumbency.append(findCount(str(newdf.iloc[[x]].candidate3[x])))
```

```python
newdf["candidate1_incumbency"] = c1_incumbency
newdf["candidate2_incumbency"] = c2_incumbency
newdf["candidate3_incumbency"] = c3_incumbency
```

```python
list_of_incumbent_winners = {}
for x in range(len(newdf)):
    c1_votes = newdf.iloc[[x]].candidate1_votes[x]
    c2_votes = newdf.iloc[[x]].candidate2_votes[x]
    c3_votes = newdf.iloc[[x]].candidate3_votes[x]
    if (c1_votes == None): 
        c1_votes = 0
    if (c2_votes == None): 
        c2_votes = 0
    if (c3_votes == None): 
        c3_votes = 0
    
    if (c3_votes > c2_votes and c3_votes > c1_votes):
        winner = newdf.iloc[[x]].candidate3_party[x]
    elif (c2_votes > c1_votes and c2_votes > c3_votes):
        winner = newdf.iloc[[x]].candidate2_party[x]
    else:
        winner = newdf.iloc[[x]].candidate1_party[x]

    list_of_incumbent_winners[str(newdf.iloc[[x]].year[x]) + str(newdf.iloc[[x]].state[x]) + str(newdf.iloc[[x]].district[x])] = winner
```

```python
c1p = []
c2p = []
c3p = []
winners = []
winpar = []
for c in range(len(newdf)):
    c1=0
    c2=0
    c3=0
    winner = ""
 
    if (newdf.iloc[[c]].candidate1_votes[c] != None):
        c1 = newdf.iloc[[c]].candidate1_votes[c]
    else:
        c1=0
        
        
    if (newdf.iloc[[c]].candidate2_votes[c] != None):
        c2 = newdf.iloc[[c]].candidate2_votes[c] 
    else:
        c1=0
        
    if (newdf.iloc[[c]].candidate3_votes[c] != None):
        c3 = newdf.iloc[[c]].candidate3_votes[c]  
    else:
        c1=0
    if (c3 > c2 and c3 > c1):
        winner = str(newdf.iloc[[c]].candidate3[c]) + " (" + str(newdf.iloc[[c]].candidate3_party[c]) + ")"
        winpar.append(str(newdf.iloc[[c]].candidate3_party[c]))
    elif (c2 > c3 and c2 > c1):
        winner = str(newdf.iloc[[c]].candidate2[c]) + " (" + str(newdf.iloc[[c]].candidate2_party[c]) + ")"
        winpar.append(str(newdf.iloc[[c]].candidate2_party[c]))
    else:
        winner = str(newdf.iloc[[c]].candidate1[c]) + " (" + str(newdf.iloc[[c]].candidate1_party[c]) + ")"
        winpar.append(str(newdf.iloc[[c]].candidate1_party[c]))
        
    total = c1+c2+c3
    c1p.append(c1/(total))
    c2p.append(c2/(total))
    c3p.append(c3/(total))
    winners.append(winner)
    
```

```python
def convertPartyToCoded(party):
    if party == None or party ==0:
        return -1
    if party.lower() == "republican":
        return 1
    elif party.lower() == "democrat":
        return 2
    elif party.lower() == "independent":
        return 3
    else:
        return 0
```

```python
newdf["winning_party_code"] = newdf["Winning Party"].apply(convertPartyToCoded)
newdf["president_party_code"] = newdf["presidents_party"].apply(convertPartyToCoded)
newdf["incumbent_party_code"] = newdf["incumbent_party"].apply(convertPartyToCoded)
newdf["candidate1_party_code"] = newdf["candidate1_party"].apply(convertPartyToCoded)
newdf["candidate2_party_code"] = newdf["candidate2_party"].apply(convertPartyToCoded)
newdf["candidate3_party_code"] = newdf["candidate3_party"].apply(convertPartyToCoded)
```
